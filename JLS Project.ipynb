# Recherche des fichiers EAF

```python
import os

def get_all_eaf_files(folder_path):
    """
    Parcours un dossier et ses sous-dossiers
    et retourne la liste de tous les fichiers .eaf
    """
    eaf_files = []

    for root, dirs, files in os.walk(folder_path):
        for file in files:
            if file.lower().endswith(".eaf"):
                full_path = os.path.join(root, file)
                eaf_files.append(full_path)
    
    return eaf_files


# Exemple d'utilisation :
folder = r"C:\Users\Ewan\Desktop\Projet JAPON JSL\corpus jls"
eaf_list = get_all_eaf_files(folder)

print("Fichiers EAF trouvÃƒÆ’Ã†â€™Ãƒâ€ Ã¢â‚¬â„¢ÃƒÆ’Ã¢â‚¬Â ÃƒÂ¢Ã¢â€šÂ¬Ã¢â€žÂ¢ÃƒÆ’Ã†â€™ÃƒÂ¢Ã¢â€šÂ¬Ã‚Â ÃƒÆ’Ã‚Â¢ÃƒÂ¢Ã¢â‚¬Å¡Ã‚Â¬ÃƒÂ¢Ã¢â‚¬Å¾Ã‚Â¢ÃƒÆ’Ã†â€™Ãƒâ€ Ã¢â‚¬â„¢ÃƒÆ’Ã‚Â¢ÃƒÂ¢Ã¢â‚¬Å¡Ã‚Â¬Ãƒâ€šÃ‚Â ÃƒÆ’Ã†â€™Ãƒâ€šÃ‚Â¢ÃƒÆ’Ã‚Â¢ÃƒÂ¢Ã¢â€šÂ¬Ã…Â¡Ãƒâ€šÃ‚Â¬ÃƒÆ’Ã‚Â¢ÃƒÂ¢Ã¢â€šÂ¬Ã…Â¾Ãƒâ€šÃ‚Â¢ÃƒÆ’Ã†â€™Ãƒâ€ Ã¢â‚¬â„¢ÃƒÆ’Ã¢â‚¬Â ÃƒÂ¢Ã¢â€šÂ¬Ã¢â€žÂ¢ÃƒÆ’Ã†â€™Ãƒâ€šÃ‚Â¢ÃƒÆ’Ã‚Â¢ÃƒÂ¢Ã¢â€šÂ¬Ã…Â¡Ãƒâ€šÃ‚Â¬ÃƒÆ’Ã¢â‚¬Â¦Ãƒâ€šÃ‚Â¡ÃƒÆ’Ã†â€™Ãƒâ€ Ã¢â‚¬â„¢ÃƒÆ’Ã‚Â¢ÃƒÂ¢Ã¢â‚¬Å¡Ã‚Â¬Ãƒâ€¦Ã‚Â¡ÃƒÆ’Ã†â€™ÃƒÂ¢Ã¢â€šÂ¬Ã…Â¡ÃƒÆ’Ã¢â‚¬Å¡Ãƒâ€šÃ‚Â©s :")
for f in eaf_list:
    print(f)
```

# Nettoyage des tokens

```python
# Nettoyage: chevrons + split '+' + filtre segments bruyants
import re
import unicodedata

def _drop_angle_and_inside(text):
    """
    Supprime tout ce qui est entre < > (chevrons), de facon robuste.
    """
    text = str(text)
    if '>' in text and '<' not in text:
        return text.split('>', 1)[1]
    if '<' in text and '>' not in text:
        return text.split('<', 1)[0]
    if '>' in text and '<' in text and text.find('>') < text.find('<'):
        return text.split('>', 1)[1].split('<', 1)[0]
    text = re.sub(r'<[^>]*>', '', text)
    text = text.replace('>', '')
    return text


def _clean_segment(seg):
    s = str(seg).strip()
    s = unicodedata.normalize('NFKC', s)
    s = s.lower()
    # PT colon handling: keep locus-id, or convert to pt(...).
    if re.fullmatch(r"(?i)^pt[123]?:[lrfub]$", s):
        return s
    m = re.fullmatch(r"(?i)^pt([123]?):(.+)$", s)
    if m:
        suffix = m.group(2).strip()
        if not re.fullmatch(r"(?i)[lrfub]", suffix):
            prefix = f"pt{m.group(1)}" if m.group(1) else "pt"
            return f"{prefix}({suffix})"

    # PT glued tokens: ptX -> pt+X (forces split)
    m = re.fullmatch(r"(?i)^pt([123])?([\u4e00-\u9fff\u3041-\u3096\u30a1-\u30fa].+)$", s)
    if m:
        digit = m.group(1) or ""
        rest = m.group(2)
        return f"pt{digit}+{rest}"
    if not s:
        return ''
    if s.startswith('??'):
        return 'UNKNOWN'

    # --------------------------------------------------
    # 1) Normalisation ponctuation fullwidth
    # --------------------------------------------------
    s = s.replace('\uFF1A', ':').replace('\uFF1B', ';').replace('\uFF0F', '/')
    s = s.replace('\uFF0F', '/').replace('\uFF0B', '+')
    # --------------------------------------------------
    # 1b) If '>' or '<' is isolated: a>b -> b, a<b -> a
    # --------------------------------------------------
    if '>' in s and '<' not in s:
        s = s.split('>')[-1].strip()
    if '<' in s and '>' not in s:
        s = s.split('<')[0].strip()

    # --------------------------------------------------
    # 2) Normalisation CL / GES (toutes variantes -> cl: / ges:)
    # --------------------------------------------------
    s = re.sub(r'(?i)\bcl\s*[:;/]\s*', 'cl:', s)
    s = re.sub(r'(?i)\bges\s*[:;/]\s*', 'ges:', s)
    s = re.sub(r'(?i)\baw\s*[:;/]\s*', 'aw:', s)
    s = re.sub(r'(?i)\bD\s*[:;/]\s*', 'D:', s)

    # --------------------------------------------------
    # 3) Nettoyage canaux main (LH / RH)
    # --------------------------------------------------
    s = re.sub(r'(?i)(^|[^A-Za-z])(?:rh|lh|R|L)\s*[:]\s*', r'\1', s)
    s = re.sub(r'(?i)/[LR]H\(([^)]*)\)', r'\1', s)
    s = re.sub(r'(?i)/[LR]H\b', '', s)
    s = re.sub(r'(?i)\(\s*(?:rh|lh)\s*\)', '', s)
    s = re.sub(r'(?i)\b(?:rh|lh)\b', '', s)

    # --------------------------------------------------
    # 3b) Nettoyage x2/x3 etc (y compris ×2)
    # --------------------------------------------------
    s = re.sub(r'(?i)[x\u00D7]\s*\d+', '', s)

    # --------------------------------------------------
    # 3c) If 'A/pt(...)' or 'A+pt(...)': keep pt(...)
    #     If 'A/pt' or 'A+pt': assign A to pt(A)
    # --------------------------------------------------
    if (('/' in s) or ('+' in s)) and re.search(r'(?i)\bpt[123]?\b', s):
        parts = [p for p in re.split(r'[+/]', s) if p]
        pt_idx = next((i for i, p in enumerate(parts) if re.search(r'(?i)\bpt[123]?\b', p)), None)
        if pt_idx is not None:
            pt_part = parts[pt_idx].strip()
            if not re.search(r'(?i)\bpt[123]?\s*\([^)]*\)', pt_part):
                other = next((p.strip() for i, p in enumerate(parts) if i != pt_idx and p.strip()), '')
                if other:
                    s = f"pt({other})"
            else:
                m = re.search(r'(?i)\bpt[123]?\s*\([^)]*\)', pt_part)
                if m:
                    s = m.group(0)

    # --------------------------------------------------
    # 3d) UNKNOWN/un avec / ou + : unknown(A) si A est present
    # --------------------------------------------------
    if (('/' in s) or ('+' in s)) and re.search(r'(?i)\b(?:un|unknown)\b', s):
        parts = [p for p in re.split(r'[+/]', s) if p]
        unk_idx = next((i for i, p in enumerate(parts) if re.search(r'(?i)\b(?:un|unknown)\b', p)), None)
        if unk_idx is not None:
            unk_part = parts[unk_idx].strip()
            if not re.search(r'(?i)\b(?:un|unknown)\s*\([^)]*\)', unk_part):
                other = next((p.strip() for i, p in enumerate(parts) if i != unk_idx and p.strip()), '')
                if other:
                    return f"unknown({other})"
            return 'unknown'

    # --------------------------------------------------
    # 4) Slashs: enlever aux extremites, remplacer au milieu par '+'
    # --------------------------------------------------
    s = re.sub(r'^/+', '', s)
    s = re.sub(r'/+$', '', s)
    s = re.sub(r'/+', '+', s)

    # --------------------------------------------------
    # 4a) Si un '+' subsiste, nettoyer chaque segment separement
    # --------------------------------------------------
    if '+' in s:
        parts = [p for p in s.split('+') if p]
        pt_idx = next((i for i, p in enumerate(parts) if re.search(r'(?i)^pt[123]?\b', p.strip())), None)
        if pt_idx is not None:
            pt_part = parts[pt_idx].strip()
            if not re.search(r'(?i)^pt[123]?\s*\([^)]*\)', pt_part):
                other = next((p.strip() for i, p in enumerate(parts) if i != pt_idx and p.strip()), '')
                if other:
                    return f"pt({other})"
        keep_nod = any(p.strip().lower().startswith('pt') for p in parts)
        cleaned = []
        for p in parts:
            if keep_nod and p.strip().lower() == 'nod':
                cleaned.append('nod')
                continue
            cleaned.append(_clean_segment(p))
        cleaned = [c for c in cleaned if c]
        if 'UNKNOWN' in cleaned:
            return 'UNKNOWN'
        joined = '+'.join(cleaned)
        # Après NFKC, on préfère réparer les parenthèses ASCII
        if joined.count('(') > joined.count(')'):
            joined += ')' * (joined.count('(') - joined.count(')'))
        # (optionnel) si jamais il reste des fullwidth
        if joined.count('\uFF08') > joined.count('\uFF09'):
            joined += '\uFF09' * (joined.count('\uFF08') - joined.count('\uFF09'))
        return joined

    # --------------------------------------------------
    # 4b) Unknown: remplace tout segment "un" par UNKNOWN
    #     sauf si "un" est suivi d'une lettre ou d'un chiffre
    # --------------------------------------------------
    if re.search(r'(?i)(?<![A-Za-z0-9])un(?![A-Za-z0-9])', s):
        return 'UNKNOWN'

    # --------------------------------------------------
    # 5) Coupe locale sur marqueurs bruyants
    #    -> on garde ce qu'il y a AVANT
    #    -> MAIS on ne coupe PAS apres une ')'
    # --------------------------------------------------
    # --------------------------------------------------
    # 5) Nettoyage local (hold) + mouthing m:
    #     - enlever hold
    #     - supprimer le segment (m:...)
    # --------------------------------------------------
    s = re.sub(r'(?i)\(\s*hold\s*\)', '', s)
    s = re.sub(r'(?i)\bhold\b', '', s)
    s = re.sub(r'\(\s*m:[^)]*\)', '', s)

    # --------------------------------------------------
    # 5b) Enlever (rep) / (omit...) en gardant le reste
    # --------------------------------------------------
    s = re.sub(r'(?i)\(\s*rep\s*\d*\s*\)?', '', s)
    s = re.sub(r'(?i)\brep\s*\d+\b', '', s)
    s = re.sub(r'(?i)\b\d+\s*rep\b', '', s)
    s = re.sub(r'(?i)\brep\b', '', s)
    s = re.sub(r'(?i)\bqm\b', '', s)
    s = re.sub(r'(?i)\(\s*omit[^)]*\)', '', s)

    # --------------------------------------------------
    # 5b2) Normalisation des variantes pt / dw
    # --------------------------------------------------
    s = re.sub(r'(?i)\b(pt[123]?)\s*[cn]\b', r'\1', s)
    s = re.sub(r'(?i)\b(pt[123]?)\s*dw\b', r'\1(dw)', s)
    s = re.sub(r'(?i)\b(pt[123]?)\s*dw[:/].*', r'\1(dw)', s)
    s = re.sub(r'(?i)\b(pt[123]?)\s*:\s*(\d+)\b', r'\1(\2)', s)

    # --------------------------------------------------
    # 5c) Supprimer nod/qm + parentheses seules/orphelines
    # --------------------------------------------------
    if re.fullmatch(r'(?i)(?:nod|qm)', s):
        return ''
    s = re.sub(r'^\(+', '', s)
    s = re.sub(r'\)+$', '', s)
    s = s.strip()
    if re.fullmatch(r'[()\[\]]+', s):
        return ''
    if re.fullmatch(r'[\(\[]\s*[\)\]]', s):
        return ''
    if re.fullmatch(r'[\(\[]\s*[\]\)]', s):
        return ''

    # --------------------------------------------------
    # 5c) Supprimer les numeraux et parentheses seules
    # --------------------------------------------------
    if re.fullmatch(r'\d+\([^)]*\)', s):
        return ''
    if re.match(r'^\d', s):
        return ''
    if re.fullmatch(r'[()]+', s):
        return ''

    # --------------------------------------------------
    # 6) Coupe locale sur marqueurs bruyants
    #    -> on garde ce qu'il y a AVANT
    #    -> MAIS on ne coupe PAS apres une ')'
    # --------------------------------------------------
    CUT_PATTERNS = [
        r'(?i)\bcl:',
        r'(?i)\bges:',
        r'(?i)\baw:',
        r'(?i)\bD:',
        r'\u53E3\u578B:',
        r'(?i)\bfal\b',
        r'(?i)\bfal:',
        r'(?i)\bNMM\b',
        r'(?i)\brep2?\b',
    ]

    # zone autorisee pour la coupe : avant le premier ')'
    paren_end = s.find(')')
    search_zone = s if paren_end == -1 else s[:paren_end]

    for pat in CUT_PATTERNS:
        m = re.search(pat, search_zone)
        if m:
            s = s[:m.start()].strip()
            break

    if not s:
        return ''

    # --------------------------------------------------
    # 6) Deparenthesage FINAL si le token est EXACTEMENT "(X)"
    #    (et seulement si X est simple)
    # --------------------------------------------------
    m = re.fullmatch(r'\(([^()]+)\)', s)
    if m:
        inner = m.group(1).strip()
        if inner and ':' not in inner and ' ' not in inner:
            s = inner

    m = re.search(r'(?i)\bpt\s*\([^)]*\)', s)
    if m and '/' in s:
        s = m.group(0)
    # Final cleanup for empty/mismatched parentheses
    s = re.sub(r'\(\s*\)', '', s)
    s = re.sub(r'\[\s*\]', '', s)
    s = re.sub(r'\uFF08\s*\uFF09', '', s)
    if re.fullmatch(r'[\(\[]\s*[\)\]]', s):
        return ''
    if re.fullmatch(r'[\(\[]\s*[\]\)]', s):
        return ''
    if s.count('(') > s.count(')'):
        s = s + (')' * (s.count('(') - s.count(')')))
    if s.count('\uFF08') > s.count('\uFF09'):
        s = s + ('\uFF09' * (s.count('\uFF08') - s.count('\uFF09')))
    return s


def _smart_split_plus(text):
    s = str(text)
    parts = []
    buf = []
    paren = 0
    angle = 0
    for ch in s:
        if ch == '(':
            paren += 1
        elif ch == ')' and paren > 0:
            paren -= 1
        elif ch == '<':
            angle += 1
        elif ch == '>' and angle > 0:
            angle -= 1
        if ch == '+' and paren == 0 and angle == 0:
            parts.append(''.join(buf))
            buf = []
        else:
            buf.append(ch)
    parts.append(''.join(buf))
    return parts


# ======================================================
# PIPELINE COMPLET
# ======================================================

tab['word'] = tab['word'].apply(_drop_angle_and_inside)
tab['word'] = tab['word'].astype(str).str.replace('\uFF0B', '+')
tab['word'] = tab['word'].apply(_smart_split_plus)
tab = tab.explode('word').reset_index(drop=True)
tab['pos'] = tab['pos'].astype(str).apply(lambda x: unicodedata.normalize('NFKC', x))
tab = tab[tab['pos'].str.contains(r'(名詞|動詞|代名詞)', regex=True, na=False)]
tab['word'] = tab['word'].apply(_clean_segment)

# Securite finale
tab = tab[~tab['word'].astype(str).str.match(r'(?i)^(?:aw:|D:)$', na=False)]
tab = tab[tab['word'].astype(str).str.len().gt(0)]

```

```python
# Regroupe les mots en phrases par tier selon un seuil d'ecart (ms)
gap_ms = 500
tab = tab.sort_values(['tier', 'start']).reset_index(drop=True)
tab['start'] = pd.to_numeric(tab['start'], errors='coerce')
tab['end'] = pd.to_numeric(tab['end'], errors='coerce')
# Merge rows with same tier/start/end before sentence grouping
group_cols = ['tier', 'start', 'end']
agg = {
    'word': lambda s: '/'.join(s.astype(str)),
    'pos': lambda s: '/'.join(s.astype(str)),
}
for c in tab.columns:
    if c not in group_cols + ['word', 'pos']:
        agg[c] = 'first'
tab = (tab.sort_values(group_cols)
          .groupby(group_cols, as_index=False, sort=False)
          .agg(agg))
prev_end = tab.groupby('tier')['end'].shift(1)
prev_word = tab.groupby('tier')['word'].shift(1)
tab['gap'] = tab['start'] - prev_end
def _is_boundary_word_strong(value):
    s = str(value).strip().lower()
    markers = ("stop", "owari", "\u7d42\u308f\u308b", "\u7d42\u308f\u308a")
    return any(m in s for m in markers)
def _is_boundary_word_weak(value):
    s = str(value).strip().lower()
    return "?" in s
def _is_boundary_word_any(value):
    return _is_boundary_word_strong(value) or _is_boundary_word_weak(value)
tab['prev_boundary'] = prev_word.apply(_is_boundary_word_strong)
tab['prev_boundary_weak'] = prev_word.apply(_is_boundary_word_weak)
tab['separation'] = tab['prev_boundary'].astype(int)
tab['boundary_strength'] = tab['prev_boundary'].map({True: 2, False: 0})
tab.loc[~tab['prev_boundary'] & tab['prev_boundary_weak'], 'boundary_strength'] = 1
tab['new_sentence'] = tab['gap'].fillna(0).ge(gap_ms) | tab['prev_boundary'] | tab['prev_boundary_weak']
tab['sentence_id'] = tab.groupby('tier')['new_sentence'].cumsum()
# Retire les tokens boundary apres la segmentation
tab = tab[~tab['word'].apply(_is_boundary_word_strong)]
def _merge_fs_tokens(tokens, pos):
    merged_tokens = []
    merged_pos = []
    i = 0
    while i < len(tokens):
        t = tokens[i]
        if isinstance(t, str) and t.lower().startswith('fs:'):
            parts = []
            keep_pos = pos[i] if i < len(pos) else None
            while i < len(tokens) and isinstance(tokens[i], str) and tokens[i].lower().startswith('fs:'):
                parts.append(tokens[i][3:])
                i += 1
            merged = ''.join(parts)
            if merged:
                merged_tokens.append(merged)
                if pos:
                    merged_pos.append(keep_pos)
            continue
        merged_tokens.append(t)
        if pos:
            merged_pos.append(pos[i])
        i += 1
    return merged_tokens, merged_pos

phrases = (tab.groupby(['tier', 'sentence_id'])
           .agg(file=('file','first'), start=('start','min'), end=('end','max'),
                tokens=('word', list), pos=('pos', list),
                separation=('separation','max'), boundary_strength=('boundary_strength','max'))
           .reset_index())
phrases[['tokens', 'pos']] = phrases.apply(
    lambda r: pd.Series(_merge_fs_tokens(r['tokens'], r['pos'])), axis=1
)
def _clean_tokens_list(tokens):
    cleaned = []
    for t in tokens:
        c = _clean_segment(t)
        if not c:
            continue
        # Ultimate fallback: pt/xxx or pt+xxx -> pt(xxx)
        if (('/' in c) or ('+' in c)) and re.search(r'(?i)\bpt[123]?\b', c) and not re.search(r'(?i)\bpt[123]?\s*\([^)]*\)', c):
            parts = [p.strip() for p in re.split(r'[+/]', c) if p.strip()]
            pt_idx = next((i for i, p in enumerate(parts) if re.fullmatch(r'(?i)pt[123]?\b', p)), None)
            if pt_idx is not None:
                other = next((p for i, p in enumerate(parts) if i != pt_idx and p), '')
                if other:
                    c = f"pt({other})"
        if ('/' in c) or ('+' in c):
            m = re.search(r'(?i)\bpt[123]?\s*\([^)]*\)', c)
            if m:
                c = m.group(0)
        # Final pass: remove empty ()/[] left after filtering
        c = re.sub(r'[\(\uff08]\s*[\)\uff09]', '', c)
        c = re.sub(r'[\[\uff3b]\s*[\]\uff3d]', '', c)
        c = c.strip()
        if c and re.fullmatch(r'[\(\[]\s*[\)\]\(\[]?\s*[\)\]]?', c):
            continue
        if c:
            cleaned.append(c)
    return cleaned
phrases['tokens'] = phrases['tokens'].apply(_clean_tokens_list)
phrases = phrases.sort_values(['file', 'start', 'end']).reset_index(drop=True)
phrases
```

```python
import ast
import json
import pandas as pd
from roles_parser import parse_all_clauses

def _parse_tokens(value):
    if value is None or str(value).strip() == "":
        return []
    txt = str(value).strip()
    try:
        return json.loads(txt)
    except Exception:
        try:
            return ast.literal_eval(txt)
        except Exception:
            return []

def _parse_dict(value):
    if value is None or str(value).strip() == "":
        return None
    txt = str(value).strip()
    try:
        return json.loads(txt)
    except Exception:
        try:
            return ast.literal_eval(txt)
        except Exception:
            return None

df_clauses = pd.read_csv("data/clause_json.csv")

clauses = []
for _, row in df_clauses.iterrows():
    clause_flags = _parse_dict(row.get("clause_flags")) or {}
    boundary_strength = row.get("boundary_strength")
    if isinstance(boundary_strength, str) and boundary_strength.strip():
        clause_flags["boundary_strength"] = boundary_strength.strip()

    clauses.append({
        "tokens": _parse_tokens(row.get("tokens")),
        "split": _parse_dict(row.get("split")),
        "boundary_seen": bool(row.get("boundary_seen")),
        "clause_flags": clause_flags,
        "clause_index": row.get("clause_index"),
        "sentence_id": row.get("sentence_id"),
        "file": row.get("file"),
        "tier": row.get("tier"),
        "start": row.get("start"),
        "end": row.get("end"),
    })

parsed = parse_all_clauses(clauses)

def _candidate_conf(candidates, source, default=0.0):
    for item in candidates or []:
        if item.get("source") == source:
            return item.get("conf", default)
    return default

def _boundary_strength(debug):
    strength = debug.get("boundary_reset")
    return strength if strength in {"strong", "weak"} else None

rows = []
for clause_id, clause in enumerate(parsed):
    action = clause.get("action", {})
    agent = clause.get("agent", {})
    patient = clause.get("patient", {})
    debug = clause.get("debug", {})
    candidates = debug.get("agent_candidates") or []
    rows.append({
        "clause_id": clause_id,
        "action_value": action.get("value"),
        "action_conf": action.get("conf"),
        "action_rule": action.get("rule"),
        "agent_value": agent.get("value"),
        "agent_conf": agent.get("conf"),
        "agent_rule": agent.get("rule"),
        "patient_value": patient.get("value"),
        "patient_conf": patient.get("conf"),
        "patient_rule": patient.get("rule"),
        "needs_ml_agent": debug.get("needs_ml_agent"),
        "needs_ml_patient": debug.get("needs_ml_patient"),
        "agent_candidates": debug.get("agent_candidates"),
        "patient_candidates": debug.get("patient_candidates"),
        "topic_conf": _candidate_conf(candidates, "TOPIC", 0.0),
        "last_agent_conf": _candidate_conf(candidates, "LAST_AGENT", 0.0),
        "last_patient_conf": _candidate_conf(candidates, "LAST_PATIENT", 0.0),
        "unknown_conf": _candidate_conf(candidates, "UNKNOWN", 0.10),
        "has_verb": debug.get("verb_index", -1) != -1,
        "verb_index": debug.get("verb_index"),
        "noun_count": debug.get("noun_count"),
        "postverb_is_pt": debug.get("postverb_is_pt"),
        "postverb_pt_resolved": debug.get("pt_resolved"),
        "pt_is_bare": debug.get("r13_reason") == "PT_BARE_NONBLOCKING",
        "boundary_strength": _boundary_strength(debug),
    })

df_parsed = pd.DataFrame(rows)

# --- contexte (2 clauses avant + clause courante, même file) ---
tokens_df = pd.read_csv("data/jsl_tokens_train.csv")
tokens_df["start"] = pd.to_numeric(tokens_df["start"], errors="coerce")
tokens_df["end"] = pd.to_numeric(tokens_df["end"], errors="coerce")

df_clauses["start"] = pd.to_numeric(df_clauses["start"], errors="coerce")
df_clauses["end"] = pd.to_numeric(df_clauses["end"], errors="coerce")
df_clauses["_idx"] = range(len(df_clauses))
df_clauses["_window_start"] = df_clauses["start"]

for file_name, group in df_clauses.groupby("file", sort=False):
    group_sorted = group.sort_values("start")
    idxs = group_sorted["_idx"].tolist()
    starts = group_sorted["start"].tolist()
    for i, clause_idx in enumerate(idxs):
        window_start = starts[max(0, i - 2)]
        df_clauses.loc[df_clauses["_idx"] == clause_idx, "_window_start"] = window_start

contexts = []
for _, row in df_clauses.iterrows():
    file_name = row.get("file")
    w_start = row.get("_window_start")
    c_end = row.get("end")
    if pd.isna(file_name) or pd.isna(w_start) or pd.isna(c_end):
        contexts.append("")
        continue
    mask = (
        (tokens_df["file"] == file_name)
        & (tokens_df["start"] >= w_start)
        & (tokens_df["end"] <= c_end)
    )
    words = tokens_df.loc[mask, "word"].astype(str).tolist()
    contexts.append("/".join(words))

df_parsed["contexte"] = contexts

# --- trivial_unknown filter ---
trivial_unknown = (
    (df_parsed["topic_conf"] == 0)
    & (df_parsed["last_agent_conf"] == 0)
    & (df_parsed["last_patient_conf"] == 0)
)

df_parsed.loc[trivial_unknown, "needs_ml_agent"] = False
df_parsed.loc[trivial_unknown, "needs_ml_patient"] = False

df_parsed.to_csv("parsed_clauses_full.csv", index=False)

df_for_annotation = df_parsed[
    (df_parsed["needs_ml_agent"] == True) | (df_parsed["needs_ml_patient"] == True)
].copy()
df_for_annotation.to_csv("clauses_for_annotation.csv", index=False)
```

```python
import ast
import json
import pandas as pd

MARGIN_TH = 0.20
TOP2_TH = 0.05
NONUNK_TH = 0.20

df = pd.read_csv("parsed_clauses_full.csv")

def _parse_candidates(value):
    if value is None or str(value).strip() == "":
        return []
    txt = str(value).strip()
    try:
        return ast.literal_eval(txt)
    except Exception:
        try:
            return json.loads(txt)
        except Exception:
            return []

def _extract_confs(cands):
    confs = {
        "TOPIC": 0.0,
        "LAST_AGENT": 0.0,
        "LAST_PATIENT": 0.0,
        "UNKNOWN": 0.0,
    }
    for item in cands or []:
        source = item.get("source")
        if source in confs:
            confs[source] = item.get("conf", 0.0)
    return confs

def _top1_top2(confs):
    vals = sorted(confs.values(), reverse=True)
    top1 = vals[0] if vals else 0.0
    top2 = vals[1] if len(vals) > 1 else 0.0
    return top1, top2

agent_confs_list = []
patient_confs_list = []

for _, row in df.iterrows():
    agent_cands = _parse_candidates(row.get("agent_candidates"))
    patient_cands = _parse_candidates(row.get("patient_candidates"))
    agent_confs_list.append(_extract_confs(agent_cands))
    patient_confs_list.append(_extract_confs(patient_cands))

agent_top1 = []
agent_top2 = []
agent_margin = []
agent_max_non_unknown = []

patient_top1 = []
patient_top2 = []
patient_margin = []
patient_max_non_unknown = []

for confs in agent_confs_list:
    top1, top2 = _top1_top2(confs)
    agent_top1.append(top1)
    agent_top2.append(top2)
    agent_margin.append(top1 - top2)
    agent_max_non_unknown.append(max(confs["TOPIC"], confs["LAST_AGENT"], confs["LAST_PATIENT"]))

for confs in patient_confs_list:
    top1, top2 = _top1_top2(confs)
    patient_top1.append(top1)
    patient_top2.append(top2)
    patient_margin.append(top1 - top2)
    patient_max_non_unknown.append(max(confs["TOPIC"], confs["LAST_AGENT"], confs["LAST_PATIENT"]))

df["agent_top1"] = agent_top1
df["agent_top2"] = agent_top2
df["agent_margin"] = agent_margin
df["agent_max_non_unknown"] = agent_max_non_unknown

df["patient_top1"] = patient_top1
df["patient_top2"] = patient_top2
df["patient_margin"] = patient_margin
df["patient_max_non_unknown"] = patient_max_non_unknown

df["needs_ml_agent_margin"] = (
    (df["agent_top2"] > TOP2_TH)
    & (df["agent_margin"] < MARGIN_TH)
    & (df["agent_max_non_unknown"] > NONUNK_TH)
)

df["needs_ml_patient_margin"] = (
    (df["patient_top2"] > TOP2_TH)
    & (df["patient_margin"] < MARGIN_TH)
    & (df["patient_max_non_unknown"] > NONUNK_TH)
)

def _build_reason(row):
    reasons = []
    if row.get("needs_ml_agent") or row.get("needs_ml_patient"):
        reasons.append("BASE_NEEDS_ML")
    if row.get("needs_ml_agent_margin"):
        reasons.append("MARGIN_AGENT")
    if row.get("needs_ml_patient_margin"):
        reasons.append("MARGIN_PATIENT")
    return "|".join(reasons)

df_margin = df[
    (df["needs_ml_agent"] == True)
    | (df["needs_ml_patient"] == True)
    | (df["needs_ml_agent_margin"] == True)
    | (df["needs_ml_patient_margin"] == True)
].copy()

df_margin["reason_expansion"] = df_margin.apply(_build_reason, axis=1)

# Optionnel: trier par min(agent_margin, patient_margin) croissant
df_margin["min_margin"] = df_margin[["agent_margin", "patient_margin"]].min(axis=1)
df_margin = df_margin.sort_values("min_margin", ascending=True).drop(columns=["min_margin"])

df_margin.to_csv("clauses_for_annotation_margin_v2.csv", index=False)
```

```python
import numpy as np
import pandas as pd
from pathlib import Path
from joblib import dump
import matplotlib.pyplot as plt
from sklearn.model_selection import StratifiedKFold
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, confusion_matrix

out_dir_bin = Path("analyse")
out_dir_bin.mkdir(parents=True, exist_ok=True)

y_unknown = (y_agent == "UNKNOWN").astype(int)

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=0)
accs = []
f1s = []
precisions = []
recalls = []
cm_total = np.zeros((2, 2), dtype=int)

for train_idx, test_idx in cv.split(X_agent, y_unknown):
    X_train = X_agent.iloc[train_idx] if hasattr(X_agent, "iloc") else X_agent[train_idx]
    X_test = X_agent.iloc[test_idx] if hasattr(X_agent, "iloc") else X_agent[test_idx]
    y_train = y_unknown.iloc[train_idx] if hasattr(y_unknown, "iloc") else y_unknown[train_idx]
    y_test = y_unknown.iloc[test_idx] if hasattr(y_unknown, "iloc") else y_unknown[test_idx]

    model = DecisionTreeClassifier(class_weight="balanced", random_state=0)
    model.fit(X_train, y_train)
    preds = model.predict(X_test)

    accs.append(accuracy_score(y_test, preds))
    prec, rec, f1, _ = precision_recall_fscore_support(y_test, preds, pos_label=1, average="binary", zero_division=0)
    precisions.append(prec)
    recalls.append(rec)
    f1s.append(f1)
    cm_total += confusion_matrix(y_test, preds, labels=[0, 1])

metrics_df = pd.DataFrame([
    {
        "accuracy_mean": float(np.mean(accs)),
        "accuracy_std": float(np.std(accs)),
        "f1_unknown_mean": float(np.mean(f1s)),
        "f1_unknown_std": float(np.std(f1s)),
        "precision_unknown_mean": float(np.mean(precisions)),
        "recall_unknown_mean": float(np.mean(recalls)),
    }
])
metrics_df.to_csv(out_dir_bin / "metrics_agent_unknown_binary.csv", index=False)

cm_df = pd.DataFrame(cm_total, index=["true_non_unknown", "true_unknown"], columns=["pred_non_unknown", "pred_unknown"])
cm_df.to_csv(out_dir_bin / "confusion_agent_unknown_binary.csv")

print("Binary CV metrics:")
print(metrics_df)
print("\nBinary confusion matrix (sum):")
print(cm_df)

final_model = DecisionTreeClassifier(class_weight="balanced", random_state=0)
final_model.fit(X_agent, y_unknown)

feat_names = list(X_agent.columns) if hasattr(X_agent, "columns") else [f"f{i}" for i in range(X_agent.shape[1])]
importances = pd.DataFrame({"feature": feat_names, "importance": final_model.feature_importances_}).sort_values("importance", ascending=False)
print("\nTop 15 importances:")
print(importances.head(15))

fig, ax = plt.subplots(figsize=(4, 4))
im = ax.imshow(cm_total, cmap="Blues")
ax.set_title("Agent UNKNOWN vs non-UNKNOWN")
ax.set_xticks([0, 1])
ax.set_yticks([0, 1])
ax.set_xticklabels(["non-UNKNOWN", "UNKNOWN"])
ax.set_yticklabels(["non-UNKNOWN", "UNKNOWN"])
ax.set_xlabel("Predicted")
ax.set_ylabel("True")
fig.colorbar(im, ax=ax)
fig.tight_layout()
fig.savefig(out_dir_bin / "fig_agent_unknown_binary.png", dpi=150)
plt.close(fig)

top10 = importances.head(10)
fig, ax = plt.subplots(figsize=(6, 4))
ax.barh(top10["feature"][::-1], top10["importance"][::-1], color="#4C78A8")
ax.set_title("Top 10 Importances (UNKNOWN binary)")
ax.set_xlabel("Importance")
fig.tight_layout()
fig.savefig(out_dir_bin / "fig_agent_unknown_importances.png", dpi=150)
plt.close(fig)

dump(final_model, out_dir_bin / "model_agent_unknown_tree.joblib")
```

```python
import numpy as np
import pandas as pd
from pathlib import Path
from joblib import dump
import matplotlib.pyplot as plt
from sklearn.model_selection import StratifiedKFold
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, classification_report

out_dir_non = Path("analyse")
out_dir_non.mkdir(parents=True, exist_ok=True)

mask = (y_agent != "UNKNOWN")
X_non = X_agent.loc[mask]
y_non = y_agent.loc[mask]

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=0)
accs = []
f1s = []
cm_total = np.zeros((3, 3), dtype=int)
oof_true = []
oof_pred = []
labels = ["TOPIC", "LAST_AGENT", "LAST_PATIENT"]

for train_idx, test_idx in cv.split(X_non, y_non):
    X_train = X_non.iloc[train_idx] if hasattr(X_non, "iloc") else X_non[train_idx]
    X_test = X_non.iloc[test_idx] if hasattr(X_non, "iloc") else X_non[test_idx]
    y_train = y_non.iloc[train_idx] if hasattr(y_non, "iloc") else y_non[train_idx]
    y_test = y_non.iloc[test_idx] if hasattr(y_non, "iloc") else y_non[test_idx]

    model = DecisionTreeClassifier(class_weight="balanced", random_state=0)
    model.fit(X_train, y_train)
    preds = model.predict(X_test)

    accs.append(accuracy_score(y_test, preds))
    f1s.append(f1_score(y_test, preds, average="macro"))
    oof_true.extend(y_test)
    oof_pred.extend(preds)
    cm_total += confusion_matrix(y_test, preds, labels=labels)

metrics_df = pd.DataFrame([
    {
        "accuracy_mean": float(np.mean(accs)),
        "accuracy_std": float(np.std(accs)),
        "macro_f1_mean": float(np.mean(f1s)),
        "macro_f1_std": float(np.std(f1s)),
    }
])
metrics_df.to_csv(out_dir_non / "metrics_agent_nonunknown.csv", index=False)

cm_df = pd.DataFrame(cm_total, index=["true_TOPIC", "true_LAST_AGENT", "true_LAST_PATIENT"],
                     columns=["pred_TOPIC", "pred_LAST_AGENT", "pred_LAST_PATIENT"])
cm_df.to_csv(out_dir_non / "confusion_agent_nonunknown.csv")

print("Non-UNKNOWN CV metrics:")
print(metrics_df)
print("\nConfusion matrix (sum):")
print(cm_df)
print("\nClassification report (OOF):")
print(classification_report(oof_true, oof_pred, labels=labels, zero_division=0))

final_model = DecisionTreeClassifier(class_weight="balanced", random_state=0)
final_model.fit(X_non, y_non)

feat_names = list(X_non.columns) if hasattr(X_non, "columns") else [f"f{i}" for i in range(X_non.shape[1])]
importances = pd.DataFrame({"feature": feat_names, "importance": final_model.feature_importances_}).sort_values("importance", ascending=False)
print("\nTop 15 importances:")
print(importances.head(15))

fig, ax = plt.subplots(figsize=(4.5, 4.5))
im = ax.imshow(cm_total, cmap="Blues")
ax.set_title("Agent non-UNKNOWN")
ax.set_xticks([0, 1, 2])
ax.set_yticks([0, 1, 2])
ax.set_xticklabels(labels, rotation=45, ha="right")
ax.set_yticklabels(labels)
ax.set_xlabel("Predicted")
ax.set_ylabel("True")
fig.colorbar(im, ax=ax)
fig.tight_layout()
fig.savefig(out_dir_non / "fig_agent_nonunknown_confusion.png", dpi=150)
plt.close(fig)

top15 = importances.head(15)
fig, ax = plt.subplots(figsize=(6, 4))
ax.barh(top15["feature"][::-1], top15["importance"][::-1], color="#4C78A8")
ax.set_title("Top 15 Importances (non-UNKNOWN)")
ax.set_xlabel("Importance")
fig.tight_layout()
fig.savefig(out_dir_non / "fig_agent_nonunknown_importances.png", dpi=150)
plt.close(fig)

dump(final_model, out_dir_non / "model_agent_nonunknown_tree.joblib")
```
